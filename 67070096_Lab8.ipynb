## 2. Context & Motivation: ทำไม Data Science อย่างเดียวไม่พอ

ในช่วงสัปดาห์ที่ 1–7 ผู้เรียนได้ฝึกทำงานกับข้อมูลจริง ตั้งแต่การสำรวจข้อมูล การทำความสะอาด การสร้าง feature ไปจนถึงการรวมข้อมูลจากหลายแหล่ง

อย่างไรก็ตาม เมื่อมองจากมุมของผู้กำหนดนโยบายหรือผู้บริหาร คำถามสำคัญมักไม่ใช่:

* “โมเดลนี้ใช้ algorithm อะไร”

แต่คือ:

* “เราควรตัดสินใจอย่างไรจากข้อมูลนี้”
* “เราจะอธิบายเหตุผลของการตัดสินใจให้สังคมเข้าใจได้หรือไม่”
* “ใครควรรับผิดชอบ หากการตัดสินใจนั้นส่งผลกระทบต่อสุขภาพประชาชน”

สัปดาห์นี้จึงเป็นจุดเปลี่ยนสำคัญของรายวิชา จาก **การทำงานกับข้อมูล** ไปสู่ **การทำความเข้าใจความหมายของความฉลาด (intelligence) และเหตุผล (reasoning)** ในระบบคอมพิวเตอร์

# 3) Conceptual Foundation: Artificial Intelligence ในหลายมิติ

## 3.1 Artificial Intelligence คืออะไร (มากกว่านิยามเดียว)

ในเชิงวิชาการ AI ไม่ได้มีนิยามเดียว แต่ขึ้นกับมุมมอง เช่น:

* มุมมองเชิงพฤติกรรม: ระบบที่แสดงพฤติกรรมคล้ายมนุษย์
* มุมมองเชิงเหตุผล: ระบบที่สามารถให้เหตุผลอย่างสมเหตุสมผล
* มุมมองเชิงเป้าหมาย: ระบบที่เลือกการกระทำเพื่อบรรลุเป้าหมายภายใต้ข้อจำกัด

สาระสำคัญคือ AI ไม่ได้หมายถึง “ความเก่ง” ของอัลกอริทึมเพียงอย่างเดียว แต่หมายถึง **กระบวนการตัดสินใจทั้งระบบ**

---

## 3.2 Narrow AI, General AI และกรอบความคาดหวังของสังคม

| แนวคิด     | คำอธิบาย                                      |
| ---------- | --------------------------------------------- |
| Narrow AI  | ระบบที่ถูกออกแบบให้ทำงานเฉพาะด้าน             |
| General AI | ระบบที่มีความสามารถเทียบเท่ามนุษย์ในหลายบริบท |

ไม่ใช่ว่า โมเดลใด ๆ ที่ “ดูฉลาด” จะสามารถตัดสินใจแทนมนุษย์ได้อย่างสมบูรณ์

---

## 3.3 วิวัฒนาการของ Narrow AI ในทางปฏิบัติ

ภายใต้กรอบของ Narrow AI สามารถแบ่งบทบาทของ AI ได้เป็นหลายรูปแบบ

| รูปแบบ                   | บทบาทหลัก                          | ตัวอย่างในบริบท PM2.5      |
| ------------------------ | ---------------------------------- | -------------------------- |
| Rule-based / Symbolic AI | ตัดสินใจตามกฎและเกณฑ์              | เกณฑ์แจ้งเตือนคุณภาพอากาศ  |
| Predictive AI            | พยากรณ์จากข้อมูลในอดีต             | โมเดลทำนายค่า PM2.5        |
| Generative AI            | สร้างข้อมูลหรือเนื้อหาใหม่         | สรุปรายงานอัตโนมัติ        |
| Narrative AI             | จัดโครงเรื่องและเหตุผลเพื่อสื่อสาร | อธิบายความเสี่ยงต่อสาธารณะ |

รูปแบบเหล่านี้ **ไม่ใช่คู่ตรงข้าม** แต่เป็นองค์ประกอบที่อาจทำงานร่วมกันในระบบเดียว

# 4) Knowledge Representation และ Symbolic AI


## 4.1 Knowledge vs Data

* Data: ข้อมูลดิบที่ยังไม่ตีความ
* Knowledge: ความรู้ที่ผ่านการตีความ เชื่อมโยง และตกผลึกเป็นกฎหรือความเข้าใจ

AI บางประเภทไม่ได้เรียนรู้จากข้อมูลจำนวนมาก แต่ใช้ **knowledge ที่มนุษย์สร้างขึ้นโดยตรง**

---

## 4.2 Knowledge Representation คืออะไร

Knowledge Representation คือกระบวนการแปลงความรู้ของโลกจริงให้อยู่ในรูปแบบที่เครื่องสามารถ:

* เก็บได้
* ตรวจสอบได้
* ให้เหตุผลต่อได้

รูปแบบที่พบบ่อย:

* IF–THEN Rules
* Logic statements
* Ontology

---

## 4.3 Rule-Based System และ Inference

ระบบที่ใช้กฎประกอบด้วย:

1. Knowledge Base
2. Inference Engine

**ตัวอย่างเชิงแนวคิด:**

* IF PM2.5 > 50 AND ลมอ่อน → ความเสี่ยงสุขภาพ = สูง
* IF ความเสี่ยงสุขภาพ = สูง → แนะนำให้หลีกเลี่ยงกิจกรรมกลางแจ้ง

ข้อดีคือความโปร่งใส ข้อจำกัดคือความไม่ยืดหยุ่นและการพึ่งพาความรู้ของผู้เชี่ยวชาญ

# 7) Reflection

ตอบคำถาม:

1. เหตุใด AI ที่แม่นยำทางสถิติ อาจไม่เหมาะกับการตัดสินใจเชิงนโยบาย
2. Knowledge Representation ช่วยเพิ่มหรือจำกัดความรับผิดชอบของมนุษย์อย่างไร

---

1. ทำไม AI แม่นทางสถิติ อาจไม่เหมาะกับนโยบาย

    เพราะ การคิดนโยบายจำเป็นต้องคำนึงถึงจริยธรรมและผลกระทบต่อสังคม ไม่ใช่แค่ความแม่นยำของค่าฝุ่น

2. Knowledge Representation ช่วยหรือจำกัดอย่างไร

    ช่วยให้ AI อธิบายเหตุผลได้ แต่ถ้าความรู้ไม่ครบหรือผิด จะทำให้การตัดสินใจผิดได้

3. Narrative AI กับความเชื่อมั่นสังคม

    การสื่อสารที่เข้าใจง่ายและโปร่งใสช่วยสร้างความเชื่อมั่น แต่ถ้าอธิบายผิดหรือเกินจริง อาจทำให้คนไม่เชื่อถือ AI
