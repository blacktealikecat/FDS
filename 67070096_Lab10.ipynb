### 3.1.1 Computer Vision คืออะไร (ในมุม Data Science)

**Computer Vision (CV)** คือสาขาของ AI ที่มุ่งให้คอมพิวเตอร์

> “เข้าใจข้อมูลจากภาพหรือวิดีโอในลักษณะที่ใกล้เคียงการรับรู้ของมนุษย์”

| มุมมอง   | Image Processing  | Computer Vision              |

> **Key idea:** Image Processing = เครื่องมือ, Computer Vision = เป้าหมาย

### 3.2.1 ภาพดิจิทัลคืออะไร

ภาพดิจิทัล = เมทริกซ์ของตัวเลข

* ภาพขาวดำ (Grayscale):  
  เมทริกซ์ 2 มิติ (H × W)
* ภาพสี (RGB):  
  เทนเซอร์ 3 มิติ (H × W × 3)

ตัวอย่างเชิงแนวคิด:

```
Pixel (x, y) = [R, G, B]
R, G, B ∈ [0, 255]
```

### 3.2.2 Spatial vs Semantic Information

| ระดับ          | ตัวอย่าง        | ความหมาย |
| -------------- | --------------- | -------- |
| Pixel-level    | ค่าความเข้ม     | แสง/สี   |
| Region-level   | texture, edge   | รูปร่าง  |
| Semantic-level | smoke, building | ความหมาย |

> ปัญหาด้าน Data Science ส่วนใหญ่อยู่ที่ **Semantic level**

### 3.2.3 ข้อจำกัดของ Image Representation

* แสง, มุมกล้อง, สภาพอากาศ → noise
* ภาพเดียวกันอาจตีความต่างกัน

## 4.4 Image Processing as Feature Engineering

มอง image processing เป็น **การสร้าง feature จากภาพ**

> Ex. Blur + Edge Detection  
> แนวคิด:
> * Blur → ลด noise
> * Edge → เน้นโครงสร้าง / ความคมของภาพ

````python
blur = cv2.GaussianBlur(gray, (7,7), 0)
edges1 = cv2.Canny(blur, 10, 20)
edges2 = cv2.Canny(blur, 50, 100)
edges3 = cv2.Canny(blur, 100, 200)

plt.figure(figsize=(20,4))

plt.subplot(1,4,1)
plt.imshow(blur, cmap="gray")
plt.title("Blurred")

plt.subplot(1,4,2)
plt.imshow(edges1, cmap="gray")
plt.title("Edges (Canny, 10, 20)")

plt.subplot(1,4,3)
plt.imshow(edges2, cmap="gray")
plt.title("Edges (Canny, 50, 100)")

plt.subplot(1,4,4)
plt.imshow(edges3, cmap="gray")
plt.title("Edges (Canny, 100, 200)")

plt.axis("off")
plt.show()
````

## 4.5 Image-derived Metrics (No Machine Learning)

เราจะสร้าง **numeric features จากภาพ**  
โดยไม่ใช้โมเดลใด ๆ

````python
# Metric 1: Sharpness (Variance of Laplacian)

def metric_sharpness(gray_img):
  return cv2.Laplacian(gray_img, cv2.CV_64F).var()

# Metric 2: Contrast (Standard Deviation)

def metric_contrast(gray_img):
  return float(np.std(gray_img))

# ทดสอบกับภาพตัวอย่าง

metric_sharpness(gray), metric_contrast(gray)
````

## 4.6 เปรียบเทียบ Metrics ระหว่างกลุ่มภาพ

คิดเหมือนการวิเคราะห์ feature ปกติใน Data Science

````python
rows = []

for p in Path("images/clear").glob("*.jpg"):
  img_clear = cv2.imread(str(p))
  img_hazy = cv2.imread(f"images/hazy/{p.stem}_1_0.2.jpg")
  g_clear = cv2.cvtColor(img_clear, cv2.COLOR_BGR2GRAY)
  g_hazy = cv2.cvtColor(img_hazy, cv2.COLOR_BGR2GRAY)
  rows.append({
    "index": p.stem,
    "clear_sharpness": metric_sharpness(g_clear),
    "clear_contrast": metric_contrast(g_clear),
    "hazy_sharpness": metric_sharpness(g_hazy),
    "hazy_contrast": metric_contrast(g_hazy)
  })

df = pd.DataFrame(rows).set_index('index').sort_index()
df.head()
````

````python
# ค่าเฉลี่ยของแต่ละกลุ่ม

df.mean()
````

````python
# ดู distribution แบบง่าย (boxplot)

for col in ["sharpness", "contrast"]:
    plt.figure(figsize=(5,4))
    data_clear = df[f'clear_{col}']
    data_hazy  = df[f'hazy_{col}']
    plt.boxplot([data_clear, data_hazy], tick_labels=["clear","hazy"])
    plt.title(f"Metric comparison: {col}")
    plt.ylabel(col)
    plt.show()
````

**Checkpoint:**

## 4.7 ชุด metrics ที่ ควรรู้จัก

ใน Data Science
> image metric = **numeric feature**  

คำถามไม่ใช่ “metric นี้ถูกไหม”  
แต่คือ  
> **metric นี้ *สื่ออะไร* และ *หลอกอะไรได้บ้าง***

### 4.7.1 Mean Intensity (Brightness)

```python
np.mean(gray)
```

**สื่ออะไร**

* ความสว่างโดยรวมของภาพ

**ใช้เมื่อ**

* แสง/หมอกควัน/ควัน ทำให้ภาพสว่างขึ้นหรือหม่นลง

**ข้อจำกัด**

* เปลี่ยนตามเวลา/แสง/กล้อง
* ไม่เกี่ยวกับ “ความคม” เลย

### 4.7.2 Intensity Range (Max − Min)

```python
gray.max() - gray.min()
```

**สื่ออะไร**

* dynamic range ของภาพ

**ข้อจำกัด**

* sensitive ต่อ outlier (pixel เดียวก็พัง)

### 4.7.3 Standard Deviation (Contrast)

```python
np.std(gray)
```

**สื่ออะไร**

* บอกระดับ “ความแตกต่างของโทนสว่าง–มืด” โดยรวมในภาพ

**ข้อจำกัด**

* ไวต่อ noise, outlier, แสง และกล้อง
* ค่า contrast สูง ไม่ได้แปลว่าภาพ “ดี” เสมอ

### 4.7.4 Entropy (ความหลากหลายของ pixel)

```python
from skimage.measure import shannon_entropy
shannon_entropy(gray)
```

**สื่ออะไร**

* ภาพมี “ความหลากหลายของ intensity” แค่ไหน

**ใช้เมื่อ**

* haze / blur → entropy มักลดลง

**ข้อดี**

* ไม่สนใจตำแหน่ง pixel
* เป็น global feature ที่ตีความได้

**ข้อจำกัด**

* ภาพ noisy อาจ entropy สูง แต่ไม่ได้ “ดี”

### 4.7.5 Edge Density

```python
edges = cv2.Canny(gray, 100, 200)
edges.mean() / 255
```

**สื่ออะไร**

* สัดส่วนของโครงสร้าง/ขอบในภาพ

**ใช้เมื่อ**

* haze / blur → edge หาย

**ข้อดี**

* เข้าใจง่าย
* เชื่อมกับ human perception

**ข้อจำกัด**

* sensitive ต่อ threshold

### 4.7.6 Gradient Magnitude (Sobel)

```python
sx = cv2.Sobel(gray, cv2.CV_64F,1,0)
sy = cv2.Sobel(gray, cv2.CV_64F,0,1)
np.mean(np.sqrt(sx**2 + sy**2))
```

**สื่ออะไร**

* ความเปลี่ยนแปลงของ intensity โดยรวม

**ข้อดี**

* smooth กว่า Canny
* threshold-free

### 4.7.7 Local Variance

```python
gray.var()
```

หรือคำนวณเป็น patch ๆ

**สื่ออะไร**

* texture / ความหยาบ

**ใช้เมื่อ**

* haze ทำให้ texture หาย

### 4.7.8 Mean Saturation (HSV)

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
np.mean(hsv[:,:,1])
```

**สื่ออะไร**

* ภาพซีด/หม่น (haze มัก saturation ต่ำ)

**ข้อดี**

* เข้าใจง่าย เชิง perception

**ข้อจำกัด**

* กล้อง/white balance มีผล

---

คำนวณ metric สำหรับทุกภาพ

```
import cv2
import numpy as np
import pandas as pd
import glob

def metric_entropy(gray_img):
    hist = cv2.calcHist([gray_img],[0],None,[256],[0,256])
    hist = hist / hist.sum()
    hist = hist[hist > 0]
    entropy = -np.sum(hist * np.log2(hist))
    return float(entropy)

results = []

for img_path in glob.glob("images/*.jpg"):
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    entropy = metric_entropy(gray)

    results.append({
        "image": img_path,
        "entropy": entropy
    })

df_metric = pd.DataFrame(results)
df_metric.head()
```
