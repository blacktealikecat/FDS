# 3) Conceptual Foundation: Logical vs Probabilistic Reasoning

## 3.1 Inductive and Deductive Reasoning in Data Science

ก่อนที่เราจะเรียนรู้ **Logical Reasoning**, **Probabilistic Reasoning** และ **Bayesian Reasoning**
จำเป็นอย่างยิ่งที่ผู้เรียนต้องเข้าใจว่า

> “เรากำลัง *ให้เหตุผล* (reason) จากอะไร → ไปสู่อะไร”

ในวิทยาการข้อมูล การให้เหตุผลหลัก ๆ มีอยู่ 2 รูปแบบสำคัญ ได้แก่

* **Deductive Reasoning (นิรนัย)**
* **Inductive Reasoning (อุปนัย)**

ซึ่งทั้งสองรูปแบบนี้เป็น “กลไกการคิด” ที่อยู่เบื้องหลังวิธี reasoning ทุกแบบที่ใช้ใน Data Science

---

### 3.1.1 Deductive Reasoning (การให้เหตุผลแบบนิรนัย)

#### ความหมาย

**Deductive Reasoning** คือการให้เหตุผลจาก

> “กฎหรือหลักการทั่วไป” → “กรณีเฉพาะ”

หากกฎตั้งต้น (premise) เป็นจริง
ผลลัพธ์ที่ได้จาก deduction จะ **ต้องเป็นจริงเสมอ** (logically guaranteed)

---

#### โครงสร้างพื้นฐานของ Deduction

1. Premise (กฎ / หลักการทั่วไป)
2. Observation (ข้อเท็จจริงเฉพาะกรณี)
3. Conclusion (ข้อสรุป)

ถ้า Premise ถูกต้อง → Conclusion จะถูกต้องโดยตรรกะ

---

#### ตัวอย่างเชิง PM2.5

**Premise (กฎ):**

> ถ้า PM2.5 มากกว่า 50 µg/m³ ต้องประกาศเตือนประชาชน

**Observation:**

> วันนี้ PM2.5 = 55 µg/m³

**Conclusion:**

> ต้องประกาศเตือนประชาชน

นี่คือ Deductive reasoning แบบสมบูรณ์

---

#### Deductive Reasoning ในรูปของโค้ด

```python
if pm25 > 50:
    alert = True
```

จุดสำคัญ:

* ไม่มีความน่าจะเป็น
* ไม่มีความคลุมเครือ
* ผลลัพธ์ถูก “บังคับ” ด้วยกฎ

---

#### จุดเด่นของ Deductive Reasoning

* อธิบายง่าย
* ตรวจสอบได้
* เหมาะกับ

  * นโยบาย (policy)
  * กฎหมาย
  * กฎความปลอดภัย
  * Rule-based systems

---

#### ข้อจำกัดของ Deductive Reasoning

* ถ้ากฎตั้งต้นผิด → ข้อสรุปผิดทั้งหมด
* ไม่รองรับความไม่แน่นอน
* ไม่สะท้อนโลกจริงที่มีความคลุมเครือ

> **Deduction ไม่ถามว่า “โอกาสเท่าไร”**
> แต่ถามว่า “เข้าเงื่อนไขหรือไม่”

---

### 3.1.2 Inductive Reasoning (การให้เหตุผลแบบอุปนัย)

#### ความหมาย

**Inductive Reasoning** คือการให้เหตุผลจาก

> “ข้อมูลหรือกรณีเฉพาะจำนวนมาก” → “ข้อสรุปทั่วไปหรือแนวโน้ม”

ข้อสรุปจาก induction:

* **ไม่รับประกันความจริง 100%**
* เป็นการประเมินความน่าจะเป็นหรือความเชื่อ

---

#### โครงสร้างพื้นฐานของ Induction

1. Observations (ข้อมูลในอดีต)
2. Pattern / Frequency (รูปแบบที่พบ)
3. Generalization (ข้อสรุปเชิงแนวโน้ม)

---

### ตัวอย่างเชิง PM2.5

จากข้อมูลย้อนหลัง 4 ปี พบว่า

> ในวันที่ลมสงบ
> ประมาณ 70% ของวันเหล่านั้น PM2.5 จะเกิน 50 µg/m³

ข้อสรุปเชิงอุปนัย:

> ถ้าวันนี้ลมสงบ มีแนวโน้มสูงที่ PM2.5 จะสูง

นี่คือ Inductive reasoning

---

#### Inductive Reasoning ในรูปของโค้ด

```python
P_high_pm_given_low_wind = 0.7
```

หรือในรูป decision:

```python
if risk_prob >= 0.6:
    alert = True
```

---

#### จุดเด่นของ Inductive Reasoning

* รองรับความไม่แน่นอน
* สอดคล้องกับโลกจริง
* เป็นพื้นฐานของ

  * Probabilistic reasoning
  * Bayesian reasoning
  * Machine Learning

---

#### ข้อจำกัดของ Inductive Reasoning

* Overgeneralization (สรุปเกินข้อมูล)
* ขึ้นกับคุณภาพข้อมูล
* อ่อนไหวต่อ bias

> **Induction ตอบคำถามว่า “น่าจะเป็นเท่าไร”**  
> ไม่ใช่ “ต้องเป็นหรือไม่”

---

### 3.1.3 เปรียบเทียบ Deductive vs Inductive Reasoning

| ประเด็น       | Deductive         | Inductive         |
| ------------- | ----------------- | ----------------- |
| จุดตั้งต้น    | กฎ                | ข้อมูล            |
| ทิศทางการคิด  | ทั่วไป → เฉพาะ    | เฉพาะ → ทั่วไป    |
| ความมั่นใจ    | 100% (ถ้ากฎจริง)  | เป็นความน่าจะเป็น |
| ความไม่แน่นอน | ไม่มี             | มี                |
| ตัวอย่างใน DS | Rule-based system | Probability / ML  |

---

### 3.1.4 ความเชื่อมโยงกับ Logical และ Probabilistic Reasoning

ในบริบทของรายวิชานี้:

* **Logical Reasoning**  
  → เป็นรูปแบบของ **Deductive Reasoning**  
  → ใช้กฎ + Logical Operators → สรุปผลแบบเด็ดขาด

* **Probabilistic / Bayesian Reasoning**  
  → เป็นรูปแบบของ **Inductive Reasoning**  
  → ใช้ข้อมูล → ประเมินความเสี่ยงและความเชื่อ

ดังนั้น เมื่อเราเขียนโค้ดแบบผสม:

```python
if not logic_pass:
    decision = "No Alert"
elif risk_prob >= threshold:
    decision = "Alert"
else:
    decision = "Monitor"
```

เรากำลังใช้:

* Deduction เพื่อกำหนด “กรอบที่ห้ามผิด”
* Induction เพื่อประเมิน “ระดับความเสี่ยงภายในกรอบนั้น”

## 3.2 Logical Reasoning and Logical Operators

จาก Section 3.1 เราได้เห็นแล้วว่า  
**Deductive Reasoning** คือการสรุปผลจาก “กฎ” ไปสู่ “กรณีเฉพาะ”

**Logical Reasoning** คือการนำ Deductive Reasoning มาทำให้เป็นรูปธรรม  
ผ่าน *ตัวดำเนินการทางตรรกศาสตร์ (Logical Operators)* เพื่อสร้างกฎที่คอมพิวเตอร์เข้าใจได้

ในงาน Data Science, Logical Reasoning ถูกใช้เป็น

* Rule-based decision
* Policy constraint
* Safety condition
* Baseline model ก่อน Machine Learning

---

### 3.2.1 Logical Operators พื้นฐาน (Primitive Operators)

Logical Operators พื้นฐานมี 3 ตัว ซึ่งถือเป็น **primitive** ของตรรกศาสตร์ทั้งหมด

| Operator | ชื่อ | สัญลักษณ์ | ความหมาย                            | Python    |
| -------- | -- |--------- | ----------------------------------- | --------- |
| NOT      | Negation | ¬A        | ปฏิเสธค่าความจริง                   | `not A`   |
| AND      | Conjunction | A ∧ B     | จริงเมื่อทุกเงื่อนไขจริง            | `A and B` |
| OR       | Disjunction | A ∨ B     | จริงเมื่ออย่างน้อยหนึ่งเงื่อนไขจริง | `A or B`  |

ตัวดำเนินการอื่น ๆ ทั้งหมดสามารถ **แปลงกลับมาเป็น 3 ตัวนี้ได้**

---

### 3.2.2 Logical Operators เชิงเหตุผล (Derived Operators)

ในงาน AI และ Data Science มักใช้ตัวดำเนินการที่ซับซ้อนขึ้น ได้แก่

* **Implication (→)** : ถ้า–แล้ว
* **Biconditional (↔)** : ก็ต่อเมื่อ
* **Exclusive OR (⊕)** : หรืออย่างใดอย่างหนึ่ง

ตัวดำเนินการเหล่านี้ช่วยให้เราสื่อ “เหตุผลเชิงนโยบาย” ได้ชัดเจนขึ้น

---

#### (1) Implication: ถ้า–แล้ว (A → B)

**ความหมาย**

> ถ้า A เป็นจริง แล้ว B ต้องเป็นจริง

หรือ

> ไม่มีกรณีใดที่ A เป็นจริง
> แต่ B เป็นเท็จ

**Truth Table**

| A | B | A → B | คำอธิบาย    |
| - | - | ----- | ----------- |
| T | T | T     | สอดคล้อง    |
| T | F | **F**     | **ขัดแย้ง** |
| F | T | T     | ไม่ขัด      |
| F | F | T     | ไม่ขัด      |

**การแปลงเป็น AND / OR / NOT**

$ A \rightarrow B \equiv (\lnot A) \lor B $


**ตัวอย่าง (PM2.5)**

> **ถ้าวันใด PM2.5 อยู่ในระดับสีแดง  
> → วันนั้นต้องถูกจัดอยู่ในกลุ่ม “ความเสี่ยงสูง”**

กำหนด

* A: `pm25_level == "Red"`
* B: `risk_level == "High"`

ตรรกะ:

$
A \rightarrow B
$

ความหมาย:

> “ไม่มีวันที่ PM2.5 เป็นสีแดง
> แต่ถูกจัดว่าไม่เสี่ยงสูง”

**Python (logical expression)**

```python
(not (pm25_level == "Red")) or (risk_level == "High")
```
Implication เป็นหัวใจของ

* Rule-based system
* Policy enforcement
* Constraint checking

---

#### (2) Biconditional: ก็ต่อเมื่อ (A ↔ B)

**ความหมาย**

> A เป็นจริง ก็ต่อเมื่อ B เป็นจริง  
> (ทั้งสองต้องมีค่าความจริงเหมือนกัน)

**Truth Table**

| A | B | A ↔ B |
| - | - | ----- |
| T | T | T     |
| T | F | F     |
| F | T | F     |
| F | F | T     |

**การแปลง**

$
A \leftrightarrow B \equiv (A \rightarrow B) \land (B \rightarrow A)
$

หรือ

```python
((not A or B) and (not B or A))
```

**ตัวอย่าง**

> จะเตือนประชาชน *ก็ต่อเมื่อ* PM2.5 สูง

---

#### (3) Exclusive OR (XOR): หรืออย่างใดอย่างหนึ่ง

**ความหมาย**

> จริงเมื่อ A หรือ B เป็นจริง “เพียงตัวเดียว”

**Truth Table**

| A | B | A ⊕ B |
| - | - | ----- |
| T | T | F     |
| T | F | T     |
| F | T | T     |
| F | F | F     |

**การแปลง**

$
A \oplus B \equiv (A \lor B) \land \lnot (A \land B)
$

```python
(A or B) and not (A and B)
```

---

### 3.2.3 Logical Reasoning ในรูป Rule-based System

เมื่อรวม Logical Operators เข้าด้วยกัน
เราจะได้ **Logical Reasoning แบบ Deductive เต็มรูป**

ตัวอย่างโค้ด:

```python
if (pm25 > 50) and (wind_speed < 2):
    decision = "Alert"
```

ลักษณะสำคัญ:

* ผลลัพธ์ “เด็ดขาด”
* ไม่มีระดับความเชื่อ
* เหมาะกับกฎที่ “ห้ามผิด”

อย่างไรก็ตาม โลกจริงไม่ได้เป็นตรรกะล้วน
จึงต้องพึ่ง **Inductive / Probabilistic Reasoning** เข้ามาเสริม

## 3.3 Probabilistic and Bayesian Reasoning (and Integration)

### 3.3.1 Probabilistic Reasoning: เหตุผลภายใต้ความไม่แน่นอน

จาก Section 3.1
**Probabilistic Reasoning** คือรูปแบบของ **Inductive Reasoning**

แทนที่จะถามว่า

> “เข้าเงื่อนไขหรือไม่”

เราถามว่า

> “มีโอกาสมากน้อยแค่ไหน”

---

### 3.3.2 Conditional Probability

แนวคิดพื้นฐานคือ

$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$

**ตัวอย่าง (PM2.5)**

* A: PM2.5 > 50
* B: ลมสงบ

> ถ้าลมสงบ โอกาสที่ PM2.5 สูงเป็นเท่าไร

```python
P_high_pm_given_low_wind = 0.7
```

---

## 3.3.3 Bayesian Reasoning: การอัปเดตความเชื่อ

Bayesian Reasoning คือการอัปเดตความเชื่อเมื่อมีข้อมูลใหม่

$
P(A|B) = \frac{P(B|A)P(A)}{P(B)}
$

โดยที่

* A = เหตุการณ์ที่เราสนใจ (เช่น PM2.5 สูง)
* B = หลักฐานใหม่ (เช่น ลมสงบ)

### องค์ประกอบ:

* **Prior**: ความเชื่อเดิม

จากข้อมูลย้อนหลังหลายปี พบว่า

> โดยเฉลี่ย  
> **30% ของวันทั้งหมด PM2.5 จะเกิน 50 µg/m³**

ดังนั้น
> $ P(\text{PM2.5 สูง}) = 0.30 $

* **Evidence**: ข้อมูลใหม่

วันนี้ตรวจพบว่า

> **ลมสงบ**

จากสถิติย้อนหลัง:

> ในวันที่ PM2.5 สูง  
  **60% ของวันเหล่านั้นเป็นวันที่ลมสงบ**

นั่นคือ
> $ P(\text{ลมสงบ} \mid \text{PM2.5 สูง}) = 0.60 $

และจากข้อมูลทั่วไป:

> วันที่ลมสงบ คิดเป็น **40% ของทุกวัน**  
> $ P(\text{ลมสงบ}) = 0.40 $

* **Posterior**: ความเชื่อหลังอัปเดต

แทนค่าในสูตร Bayes:

$
P(\text{PM2.5 สูง} \mid \text{ลมสงบ})
= \frac{0.60 \times 0.30}{0.40}
= 0.45 $

---

### การตีความ

ก่อนรู้เรื่องลม:

* ความเชื่อว่า PM2.5 จะสูง = **30%**

หลังรู้ว่าลมสงบ:

* ความเชื่อเพิ่มเป็น = **45%**

> **นี่คือ “การอัปเดตความเชื่อ”**  
> ไม่ใช่การทำนายแบบเด็ดขาด  
> แต่เป็นการ *ปรับระดับความเสี่ยง* ตาม evidence ใหม่

---

````python
# Prior
P_high_pm = 0.30

# Likelihood
P_low_wind_given_high_pm = 0.60

# Evidence
P_low_wind = 0.40

# Bayesian update
P_high_pm_given_low_wind = (
    P_low_wind_given_high_pm * P_high_pm
) / P_low_wind

P_high_pm_given_low_wind
````

สมมติว่านโยบายกำหนดว่า

> ถ้า PM2.5 ≤ 37  
> → *ไม่ต้องประเมินความเสี่ยงเชิงความน่าจะเป็น*

ขั้นตอน reasoning จริงจึงเป็น

```python
if pm25 <= 37:
    decision = "No Alert (Logic)"
else:
    if P_high_pm_given_low_wind >= 0.6:
        decision = "Alert"
    else:
        decision = "Monitor"
```

🔹 **Logic = กรอบบังคับ**  
🔹 **Bayes = อัปเดตระดับความเชื่อภายในกรอบ**

````python
# http://air4thai.pcd.go.th/webV3/#/Home
kmitl_pm_df = df[['Date', '78T', 'wind_speed_10m_max']].copy()
kmitl_pm_df.columns = ['Date', 'pm25', 'wind_speed']
kmitl_pm_df
````

## 4.2 โจทย์ (Single Integrated Scenario)

### สถานการณ์

> ระบบพิจารณาว่าจะ “ประกาศเตือน PM2.5” หรือไม่

**นโยบาย (Logical rules)**

1. ถ้า PM2.5 ≤ 20 → *ห้ามเตือน*
2. ถ้า PM2.5 > 20 → *พิจารณาต่อได้*

**การประเมินความเสี่ยง (Probabilistic)**

* หากลมสงบ → ความน่าจะเป็นที่ PM2.5 จะสูงต่อเนื่อง = 0.7
* หากลมแรง → ความน่าจะเป็น = 0.2

**เกณฑ์การตัดสินใจ**

* เตือน เมื่อ probability ≥ 0.6

---

## 4.3 Logical Layer: ตรวจสอบเงื่อนไขบังคับ


````python
kmitl_pm_df["logic_pass"] = kmitl_pm_df["pm25"] > 20
kmitl_pm_df['logic_pass'].value_counts()
````

* `True` → ไปขั้น probability
* `False` → หยุดการตัดสินใจทันที

---

## 4.4 Probabilistic Layer: ประเมินความเสี่ยง


````python
def risk_probability(row):
    if row["wind_speed"] < 12:
        return 0.7
    else:
        return 0.2

kmitl_pm_df["risk_prob"] = kmitl_pm_df.apply(risk_probability, axis=1)
kmitl_pm_df['risk_prob'].value_counts()
````

---

## 4.5 รวม Logical + Probabilistic Decision

````python
def final_decision(row):
    # Logical constraint
    if not row["logic_pass"]:
        return "No Alert (Logic)"

    # Probabilistic decision
    if row["risk_prob"] >= 0.6:
        return "Alert"
    else:
        return "Monitor"

kmitl_pm_df["decision"] = kmitl_pm_df.apply(final_decision, axis=1)
kmitl_pm_df['decision'].value_counts()
````
